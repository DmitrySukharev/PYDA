**Уровень 0**  
Придумайте матрицы, которые обладают следующими свойствами:  
a) диагональная, ортогональная, но не единичная(!):  
b) верхнетреугольная, ортогональная:  
c) нижнетреугольная, ортогональная:  
d) матрица, у которой определитель = 1, и все элементы не равны 0:  

P.S. Во всех пунктах ответом должна быть ненулевая матрица размерности не менее 2!  

```python
import numpy as np
```

```python
a = np.array([[1, 0], [0, -1]])
print(f"a) диагональная, ортогональная, но не единичная:\n{a}\n")
print(f"При проверке транспонированием должна получиться единичная матрица:\n{a @ a.transpose()}")
```

```python
# b) верхнетреугольная, ортогональная
# c) нижнетреугольная, ортогональная
# один ответ - диагональная матрица из 1 или -1

b = np.array([[-1, 0], [0, -1]])
print(f"верхнетреугольная и одновременно нижнетреугольная, ортогональная\n{b}\n")
print(f"При проверке транспонированием должна получиться единичная матрица:\n{b @ b.transpose()}")
```

```python
# матрица, у которой определитель = 1, и все элементы не равны 0
d = np.array([[3, 2], [4, 3]])
print(d)
print(f"Детерминант: {np.linalg.det(d)}")
```

Найдите собственные значения и собственные вектора матриц:  
m = np.array([[2, 2],  
[1, 3]])  

```python
m = np.array([[2, 2],
[1, 3]])
print(m)
print(f"Собственные значения: {np.linalg.eig(m)[0][0]} и {np.linalg.eig(m)[0][1]}")
print(f"Собственные векторы:\n {np.linalg.eig(m)[1]}")
# т.е. собственные значения 1 и 4, а собственные вектора-столбцы (1, 1) и (2, -1)
```
