Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
sigma = 0.5

for mu in [0.5, 1, 2, 5, 8]:
    norm_rv = sts.norm(loc=mu, scale=sigma)
#    cdf = norm_rv.cdf(x)
    pdf = norm_rv.pdf(x)
#    plt.plot(x, cdf)
    plt.plot(x, pdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
x = np.linspace(0, 5, 100)
mu = 2

for sigma in [0.2, 0.3, 0.5, 1, 2]:
    norm_rv = sts.norm(loc=mu, scale=sigma)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
x = np.linspace(0,5,100)

for lam in [0.1, 0.2, 0.5, 1, 2]:
    expon_rv = sts.expon(lam)
    pdf = expon_rv.pdf(x)
    plt.plot(x, pdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
x = np.linspace(-5,5,100)

for n in [0.1, 1, 5, 10, 10000]:
    t_rv = sts.t(n)
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```
