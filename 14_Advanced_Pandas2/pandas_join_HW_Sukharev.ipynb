**Задание 1** 

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head(3)
```

```python
users = ratings.groupby('userId').agg({'timestamp':['max', 'min'], 'rating':'count'})    # group by userId
users.loc[users['rating']['count'] > 100, 'ts_delta'] = users['timestamp']['max'] - users['timestamp']['min']  # delta calc
users.loc[users['rating']['count'] > 100].sort_values('ts_delta', ascending=False)    # the resulting table
```

```python
users.loc[users['rating']['count'] > 100]['ts_delta'].mean()
```
